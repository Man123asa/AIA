```
# Brute-Force Method for Word-Level Tokenization
# Method : Uses basic Regular Expressions (Regex) to split text on word boundaries.
# Storage : Uses hashmaps (dictionaries) to build the vocbulary index
```

```
# Conventional Regex Method
import re
class token:

  def __init__(self):
    self.encode={}
    self.decode={}

  def fit(self,text):
    tokens=re.findall(r"[\w']+|[.,!?;]" ,text.lower())
    st=sorted(list(set(tokens)))
    self.encode={t:i for i,t in enumerate(st)}
    self.decode={i:t for t,i in self.encode.items()}

  def en(self,text):
    findt=re.findall(r"[\w']+|[.,!?;]",text.lower())
    return [self.encode.get(token,-1) for token in findt]
    print(self.encode.items())

  def de(self,ids):
    return "".join([self.decode.get(i,"<?") for i in ids])

  def printdata(self,text):
    print(f"COUNT:        {len(self.encode)}")
    print()
    print(f"TOKENS:       {list(self.encode.keys())}")
    print()
    print(f"TOKEN IDs:    {list(self.encode.values())}")
    print()
```

```
# Static Test Data to test All possible test Cases


static_text=(
    "Tokenization is amazing! Learning NLP models with regex rules. "
    "नमस्ते दुनिया! Bonjour! こんにちは 世界 🚀🔥 Windows 11 vs 10."
)
tk=token()
tk.fit(static_text)
print("=== STATIC SYSTEM EVALUATION DATA ===")
tk.printdata(static_text)

# Executing tests to show encoder/decoder results directly in the PDF
print("=== STATIC TEST RUNS ===")
test_sentences=[
    "Tokenization is amazing.",
    "नमस्ते दुनिया!",
    "Bonjour!",
    "Windows 11 vs 10.",
    "Unknown tokens 🚀🔥 give -1"
]
for test in test_sentences:
    encoded=tk.en(test)
    decoded=tk.de(encoded)
    print(f"Input:   {test}")
    print(f"Encoded: {encoded}")
    print(f"Decoded: {decoded}\n")
```

```
import tiktoken
test_sentences=[
    "Tokenization is amazing.",
    "नमस्ते दुनिया!",
    "Bonjour!",
    "Windows 11 vs 10.",
    "Unknown tokens 🚀🔥 give -1"
]
print("=== STATIC TEST RUNS ===")
for test in test_sentences:
    text=test
    en=tiktoken.encoding_for_model("gpt-4o")
    tokens=en.encode(text)
    token_count=len(tokens)
    token_String=[en.decode_single_token_bytes(t).decode('utf-8',errors='replace') for t in tokens]
    decoded_string=en.decode(tokens)
    print(f"Input:   {test}")
    print(f"Encoded: {tokens}")
    print(f"Decoded: {decoded_string}\n")
```

```
!pip install sentencepiece

#Creating sample text to train the SentencePiece model
training_text="""
Retrieval-Augmented Generation (RAG) is a framework that connects an LLM to external data.
SentencePiece is an open-source tokenization library built by Google.
It treats the input text as a raw byte stream, which makes it language-independent.
Unlike traditional tokenizers, it does not require pre-tokenization spaces or rules.
Instead, it treats spaces as a normal character, often represented by an underscore.
Tokenization breaks text into subword units to eliminate out-of-vocabulary errors.
Modern language models like Llama, Mistral, and T5 rely heavily on SentencePiece.
We can train a BPE or Unigram model using this library in just a few lines of code.
"""
# Saving the text to a local file in the Colab directory
with open("training_data.txt","w",encoding="utf-8") as f:
    f.write(training_text.strip())

print("Dependencies installed and 'training_data.txt' created successfully!")
```

```
import sentencepiece as spm
# We set vocabulary size to 100 because our sample training dataset is very small.
spm.SentencePieceTrainer.train(
    input='training_data.txt',
    model_prefix='rag_tokenizer',
    vocab_size=100,
    model_type='unigram' # Options: 'unigram', 'bpe', 'char', 'word'
)
print("Tokenizer trained! Saved as 'rag_tokenizer.model' and 'rag_tokenizer.vocab'\n")
# Loading the trained model
sp=spm.SentencePieceProcessor()
sp.load('rag_tokenizer.model')

# Defining Static Test Data
test_sentences=[
    "Retrieval-Augmented Generation works with SentencePiece.",
    "Llama and Mistral use subword tokenization rules.",
    "This is a completely new sentence to test out-of-vocabulary words."
]

# Running Tokenization Tests
print("--- STARTING TOKENIZATION TESTING ---")
for idx, sentence in enumerate(test_sentences, 1):
    print(f"\nTest Case {idx}: '{sentence}'")
    token_pieces=sp.encode_as_pieces(sentence)
    print(f"Subword Pieces : {token_pieces}")
    # encode
    token_ids=sp.encode_as_ids(sentence)
    print(f"Numerical IDs   : {token_ids}")
    # decode
    decoded_text=sp.decode_ids(token_ids)
    print(f"Decoded Text    : '{decoded_text}'")
```
